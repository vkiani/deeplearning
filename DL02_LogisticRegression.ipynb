```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x = [ [10], [15], [9], [18], [3] ]
```

```python
y = [ 1, 1, 0, 1, 0 ]
```

```python
x,y = np.array(x), np.array(y)
```

```python
x
```

```python
y
```

```python
model = LogisticRegression()

model.fit(x,y)

model
```

```python
y_pred = model.predict(x)

y_pred
```

```python
y
```

```python
print("Train Accuracy of Logistic Regression model is:", accuracy_score(y, y_pred)*100)
```

```python
x_test = [ [14], [7], [2] ]
y_test = [ 1  ,  0,   0]
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Test Accuracy of Logistic Regression model is:", accuracy_score(y_test, y_pred)*100)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
def f(x):
  return np.sin(2 * np.pi * x)
```

```python
np.random.seed(9)

# generate points used to plot
x_plot = np.linspace(0, 1, 100)

# generate points and keep a subset of them
n_samples = 100
x_train = np.random.uniform(0, 1, size=n_samples)[:, np.newaxis]
y_train = f(x_train) + np.random.normal(scale=0.3, size=n_samples)[:, np.newaxis]
```

```python
import matplotlib.pyplot as plt

ax = plt.gca()
ax.plot(x_plot, f(x_plot), color='green')
ax.scatter(x_train, y_train, s=10)
```

## Polynomial Regresssion

Polynomial of Degree One

```python
x_train_poly = PolynomialFeatures(degree = 1).fit_transform(x_train)
model = LinearRegression()
model.fit(x_train_poly, y_train)

x_plot_poly = PolynomialFeatures(degree = 1).fit_transform(x_plot[:, np.newaxis])
y_plot_poly = model.predict(x_plot_poly)

ax = plt.gca()
ax.plot(x_plot, f(x_plot), color='green')
ax.scatter(x_train, y_train, s=10)
ax.plot(x_plot, y_plot_poly, color='red', linestyle='dashed')
```

Polynomial of Degree Three

```python
x_train_poly = PolynomialFeatures(degree = 3).fit_transform(x_train)
model = LinearRegression()
model.fit(x_train_poly, y_train)

x_plot_poly = PolynomialFeatures(degree = 3).fit_transform(x_plot[:, np.newaxis])
y_plot_poly = model.predict(x_plot_poly)

ax = plt.gca()
ax.plot(x_plot, f(x_plot), color='green')
ax.scatter(x_train, y_train, s=10)
ax.plot(x_plot, y_plot_poly, color='red', linestyle='dashed')
```

```python
x_train_poly = PolynomialFeatures(degree = 8).fit_transform(x_train)
model = LinearRegression()
model.fit(x_train_poly, y_train)

x_plot_poly = PolynomialFeatures(degree = 8).fit_transform(x_plot[:, np.newaxis])
y_plot_poly = model.predict(x_plot_poly)

ax = plt.gca()
ax.plot(x_plot, f(x_plot), color='green')
ax.scatter(x_train, y_train, s=10)
ax.plot(x_plot, y_plot_poly, color='red', linestyle='dashed')
plt.ylim((-1, 1))
```

## Regularization Ridge

Polynomial of Degree Ten

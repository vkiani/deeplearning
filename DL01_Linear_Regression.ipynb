```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = [ [10], [15], [9], [18], [3] ]
```

```python
y = [ 2, 3, 1.5, 3.5, 0.5 ]
```

```python
x,y = np.array(x), np.array(y)
```

```python
x
```

```python
y
```

```python
model = LinearRegression()

model.fit(x,y)

model
```

```python
model.intercept_
```

```python
model.coef_
```

```python
x
```

```python
model.predict(x)
```

```python
y
```

```python
model.score(x,y)
```

```python
x_test = [ [14], [7], [2] ]
y_test = [ 3  ,  2,   1]
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
```

```python
model.score(x_test, y_test)
```
